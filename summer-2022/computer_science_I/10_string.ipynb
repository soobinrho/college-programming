<center><b><em>Repeated-Words Detecter</em></b></center>

````python
'''Detects and prints if the input string has any repeated words.'''
import numpy as np
from collections import Counter

def print_repeated_words(str_input: str):
    '''
        1. Strips for data sanitization.
        2. Converts to all lower case.
        3. Splits str_input into words.
        4. Counts the occurence of each word.
        5. Prints how many times each word is repeated.        
    '''

    # Making a copy of str_input.
    # The reason why I use the copy instead
    # of str_input is because - at the end -
    # we are printing str_input as well.
    str_input_copy = str_input

    # 1. Strips for data sanitization.
    str_input_copy = str_input_copy.strip()

    # 2. Converts to all lower case.
    str_input_copy = str_input_copy.lower()

    # 3. Splits str_input into words.
    list_words = str_input_copy.split()

    # 4. Counts the occurence of each word.
    list_word_count = dict(Counter(list_words))

    # Which words were repeated?
    # i.e. which words were used more than once?
    list_words_repeated = list(
        filter(
            lambda name: list_word_count[name] >= 2,
            list_word_count
        )
    )

    # How many words were repeated?
    count_words_repeated = len(list_words_repeated)

    # 5. Prints how many times each word is repeated.
    if count_words_repeated != 0:
        print('')    # Printing new line.
        for word in list_words_repeated:
            count = list_word_count[word]
            print(f'{word} = {count}')

    # Printing if no words were repeated.
    # I could've just used else, but elif is for clarity.
    elif count_words_repeated == 0:
        print(f'\nNo words were repeated.')
````

````python
# Printing the title.
str_title = 'Repeated-Words Detector'
print(len(str_title) * '=')
print(str_title)
print((len(str_title) * '=') + '\n')

# Prompting for user input
str_input = input('Enter: ')
print_repeated_words(str_input)
````

<center><b><em>Variable Assignment Code Generator</em></b></center>

One way of making arrays of nouns and verbs is to manually type them. 
However, I wanted to make a bonus script such that it can be automated.

The following script first prompts for user input. 
We can then copy and paste a very long text into the input:
`Enter: a very very very very very very very very long text`

The script parses and prints the string such that 
it can be directly put into a Python script:
```python
words_source = (
    'a very very very' +
    'very very very' +
    'very very long text'
)
```

````python
'''Parses any long text for variable assignment.'''
import numpy as np
import re

def convert_to_code(
        str_input: str, 
        variable_name: str = "words_source"
    ) -> str:
    '''
        1. Removes all single quotes (') to avoid error.
        2. Splits into words.
        3. Loops through each word.        
        4. Loops until each line reaches linelength_max.    
        5. Closes the bracket.        
        6. Prints the code.
    '''

    # 1. Removes all single quotes (') to avoid error.
    str_input = re.sub(r"'", "", str_input)
    
    # 2. Splits into words.
    list_words = str_input.split()

    # The maximum length of each line.
    linelength_max = 34    # (characters)

    # Initializing the final code output.
    list_codes = np.array([f"\n{variable_name} = ("])

    # 3. Loops through each word.
    index_words = 0
    while index_words < len(list_words):

        # Initializing variables for
        # each individual line.
        each_line = ''
        len_each_line = 0

        # 4. Loops until each line reaches linelength_max.
        while (
            len_each_line <= linelength_max and
            index_words < len(list_words)
        ):
            word_added = f'{list_words[index_words]} '
            each_line += word_added
            len_each_line += len(word_added)
            index_words += 1

        # Adding '...' +
        list_codes = np.append(list_codes, f"    '{each_line}' +")

    # Removing the last line's +
    list_codes[-1] = list_codes[-1].replace(' +', '')

    # 5. Closes the bracket.
    list_codes = np.append(list_codes, ")")

    # 6. Prints the code.
    for code in list_codes:
        print(code)    
````

````python
# Printing the title.
str_title = 'Variable Assignment Code Generator'
print(len(str_title) * '=')
print(str_title)
print((len(str_title) * '=') + '\n')

# Prompting for user input.
str_input = input('Enter: ')

# Converting the input and printing the code.
convert_to_code(str_input)
````

<center><b><em>Generating Arrays of Nouns and Verbs</em></b></center>

This script compiles arrays of nouns and verbs from Dr. Leach's homework instructions.

````python
'''Converts a long string into arrays of nouns and verbs.'''
# nltk.corpus is included in Anaconda.
# However, nltk.corpus data sets need to be
# downloaded. They can be downloaded by
# opening python on your terminal and then typing:
#
# import nltk
# nltk.download()
# nltk.download_gui()
#
from nltk.corpus import wordnet as wn
from collections import Counter
import numpy as np
import re

# Removing any non-word from words_source
# for data sanitization.
words_source = re.sub(r'[^a-zA-Z ()]', '', words_source)

# Converting words_source into all lowercase.
words_source = words_source.lower()

# Splitting words_source into words.
words_source = words_source.split()

# Deleteing duplicate words.
# Converting the list into a set removes all duplicates.
words_source = set(words_source)
words_source = list(words_source)

# Initializing an array of verbs.
array_verbs = np.array([])

# Initializing an array of nouns.
array_nouns = np.array([])

# Looping through every item in words_source
# in order to categorize them either to verbs or nouns.
index_word = 0
while index_word < len(words_source):
    
    # The current word in the loop.
    word = words_source[index_word]
    
    # Filtering one or two letter words.
    # Technically, there are two letter words.
    # However, since I'm relying on nltk to
    # categorize verbs and nouns, the accuracy is not perfect.
    #
    # I've tested data with or without this filtering,
    # and I've noticed that many of the two letter words
    # are likely to be inaccurately categorized as
    # verbs or nouns -- e.g. 'be', 'no', 'or'.
    #
    # Therefore, I've decided to just filter
    # one or two letter words altogether to sacrifice
    # a part of accuracy for overall accuracy.
    if len(word) == 1 or len(word) == 2:
        index_word += 1
        continue
    
    # Filtering articles.
    is_article = False
    for article in array_articles:
        if word == article:
            is_article = True
            break
            
    if is_article == True:
        index_word += 1
        continue
        
    # Filtering prepositions.
    is_preposition = False
    for preposition in array_prepositions:
        if word == preposition:
            is_preposition = True
            break
            
    if is_preposition == True:
        index_word += 1
        continue    
        
    # The type of the word is determined using
    # the Natural Language Toolkit (NLTK) Library.
    # wn.synsets(word)[0].pos() is 
    # 'v' for verbs and 'n' for nouns.
    if len(wn.synsets(word)) > 0:
        
        # The synset could be just a synonym, and not
        # the word itself. For example, I found that
        # the word 'as' was on the nouns list.
        # This if statement can prevent it.
        synset_name = wn.synsets(word)[0].name().split('.')[0]
        if (
            not word in synset_name and
            not synset_name in word
        ):
            index_word += 1 
            continue
        
        # Adding the filtered words into nouns or verbs.
        type_of_word = wn.synsets(word)[0].pos()
        if type_of_word == 'n':
            array_nouns = np.append(array_nouns, word)
        elif type_of_word == 'v':
            array_verbs = np.append(array_verbs, word)

    index_word += 1    
````

<center><b><em>Random Sentence Generator</em></b></center>

The final program generating 20 random sentences.

````python
'''Generates a random sentence.'''
from random import randint

def generate_random_sentence() -> str:
    '''
        1. Picks an article and capitalizes the first letter.
        2. Picks a noun.
        3. Picks a verb.
        4. Picks a preposition.
        5. Picks another article.
        6. Picks another noun and ends with a period.
    '''

    # Initializing the return value.
    sentence = ""
    
    # 1. Picks an article and capitalizes the first letter.
    article = array_articles[randint(0, len(array_articles) - 1)]
    article = article.capitalize()
    sentence += f"{article} "
    
    # 2. Picks a noun.
    noun = array_nouns[
        randint(0, len(array_nouns) - 1)
    ]
    sentence += f"{noun} "
    
    # 3. Picks a verb.
    verb = array_verbs[
        randint(0, len(array_verbs) - 1)
    ]
    sentence += f"{verb} "
    
    # 4. Picks a preposition.
    preposition = array_prepositions[
        randint(0, len(array_prepositions) - 1)
    ]
    sentence += f"{preposition} "
    
    # 5. Picks another article.
    article = array_articles[
        randint(0, len(array_articles) - 1)
    ]
    sentence += f"{article} "    
    
    # 6. Picks another noun and ends with a period.    
    noun = array_nouns[randint(0, len(array_nouns) - 1)]
    sentence += f"{noun}."    
    
    return sentence
````

````python
# Printing the title.
str_title = 'Random Sentence Generator'
print(len(str_title) * '=')
print(str_title)
print((len(str_title) * '=') + '\n')

# Generate 20 random sentences.
for count in range(20):
    print(generate_random_sentence())
````
